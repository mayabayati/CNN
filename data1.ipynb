```python
train_data = pd.read_csv('/kaggle/input/fashionmnist/fashion-mnist_train.csv',sep=',')
```

```python
#creating a numpy array for train and test
traindata = np.array(train_data, dtype = 'float32')
```

```python
x_tt = traindata[:,1:]
y_tt = traindata[:,0]
```

```python
x_tt = x_tt/255
```

```python
x_tt -= 0.5
x_tt *= 2
```

```python
row = 28
col = 28
shape = (row,col,1)
```

```python
x_t,x_v,y_t,y_v= train_test_split(x_tt,y_tt,test_size = 0.2,random_state = 12345)
```

```python
x_t = x_t.reshape(x_t.shape[0],*shape)
x_v = x_v.reshape(x_v.shape[0],*shape)
y_t2 = keras.utils.to_categorical(y_t).astype('float32')
y_v2 = keras.utils.to_categorical(y_v).astype('float32')
```

```python
num_classes = 10
cnn=Sequential()
cnn.add(Conv2D(filters=32, kernel_size=(3, 3), padding="same", 
          input_shape=shape, activation='relu'))
cnn.add(Conv2D(filters=64, kernel_size=(3, 3), padding="same", activation='relu'))
cnn.add(MaxPooling2D(pool_size=(2, 2)))
cnn.add(Dropout(0.5))
cnn.add(Conv2D(filters=128, kernel_size=(3, 3), padding="same", activation='relu'))
cnn.add(Conv2D(filters=256, kernel_size=(3, 3), padding="valid", activation='relu'))
cnn.add(MaxPooling2D(pool_size=(3, 3)))
cnn.add(Dropout(0.5))
cnn.add(Flatten())
cnn.add(Dense(256))
cnn.add(LeakyReLU())
cnn.add(Dropout(0.5))
cnn.add(Dense(256))
cnn.add(LeakyReLU())
cnn.add(Dense(num_classes, activation='softmax'))
cnn.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
cnn.summary()
```

```python
data_history = cnn.fit(
    x_t,
    y_t2,
    batch_size=500,
    epochs=50,
    verbose=1,
    validation_data=(x_v,y_v2)
)
```

```python
test = pd.read_csv('/kaggle/input/datamnist/test_data.csv', sep = ',')
test_d = np.array(test, dtype='float32')
```

```python
x_tes = test_d[:,1:]/255
x_tes -= 0.5
x_tes *=2
x = x_tes.reshape(x_tes.shape[0],*shape)
p =cnn.predict_classes(x)
print(p)
```
